```python
import pandas as pd

df_train = pd.read_csv('fraudTrain.csv')
df_test = pd.read_csv('fraudTest.csv')
```

```python
# Concate the training and test data 

df = pd.concat([df_train,df_test], axis=0)
df
```

```python
# Droping the first column

df.drop('Unnamed: 0',axis='columns',inplace=True)
print(df.shape)
```

```python
# Step 1: Date and Time Column

print()
print(df.shape)
print()
print(df['trans_date_trans_time'].info())
print()
```

```python
# converting date and time column form object data_type to date and time 

df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
print(df['trans_date_trans_time'].info())
print()
```

```python
# Step 2: Extracting all important features from datetime column

# Extracting all important features from datetime column

df['years'] = df['trans_date_trans_time'].dt.year
df['months'] = df['trans_date_trans_time'].dt.month
df['weekofday'] = df['trans_date_trans_time'].dt.weekday
df['days'] = df['trans_date_trans_time'].dt.day
df['hours'] = df['trans_date_trans_time'].dt.hour
df['minutes'] = df['trans_date_trans_time'].dt.minute
df['seconds'] = df['trans_date_trans_time'].dt.second

# dividing the hours into four categories

df['time of day'] = pd.cut(df['hours'], bins=[0,6,12,18,24],labels=['Night','Morning','Afternoon','Evening'])

# Now, group the 'cc_num' column with with 'trans_date_trans_time' column and find difference 
# between each transaction of the card number.

df['time_since_last_transaction'] = df.groupby('cc_num')['trans_date_trans_time'].diff()

# Now, convert the 'time_since_last_transaction' column into seconds

df['time_since_last_transaction'] = df['time_since_last_transaction'].dt.total_seconds()
```

```python
# Drop 'trans_date_trans_time' columnAfter extracting all the important features form 
# the 'trans_date_trans_time' column

df = df.drop('trans_date_trans_time', axis=1)
print()
print(df.shape)
```

```python
# Step 3: Preprocessing of Merchant and category column


print(df['merchant'].value_counts().iloc[0:700].sum())
print()
```

```python
#Preprocessing the 'merchant' column before label encoding

# converting the merchant column into lower case

df['merchant'] = df['merchant'].str.lower()
df['category'] = df['category'].str.lower()

# Removing all the white spaces befor and after the text

df['merchant'] = df['merchant'].str.strip()
df['category'] = df['category'].str.strip()

# Replace spaces with underscore

df['merchant'] = df['merchant'].str.replace(' ', '_')
df['category'] = df['category'].str.replace(' ', '_')

# Removing commas and full stops

df['merchant'] = df['merchant'].str.replace(',', '')
df['category'] = df['category'].str.replace(',', '')

# Removing full stops

df['merchant'] = df['merchant'].str.replace('.', '')
df['category'] = df['category'].str.replace('.', '')

# Replacing - with _

df['merchant'] = df['merchant'].str.replace('-', '_')
df['category'] = df['category'].str.replace('-', '_')

# Removing all the white spaces befor and after the text

df['merchant'] = df['merchant'].str.strip()
df['category'] = df['category'].str.strip()
```

```python
from sklearn.preprocessing import LabelEncoder

le_merchant = LabelEncoder()
le_category = LabelEncoder()
print(le_merchant)
print()
print(le_category)
print()
df['merchant'] = le_merchant.fit_transform(df['merchant'])
df['category'] = le_category.fit_transform(df['category'])

columns = ['merchant','category']
print(df[columns].info())
print()
print(df.shape)
```

```python
# Step 4: Droping Irrelevant Columns


print(df.shape)
print()

# Drop the irrelevant columns

# "job" ,"cc_num"

df.drop(['first','last','street','dob','trans_num','time_since_last_transaction','time of day',"zip" ,"unix_time","merch_lat","merch_long"], axis=1, inplace=True)
print(df.shape)
df.head(1)
```

```python
# Fill the empty values

print(df.isnull().sum())
print()
print(df.info())
```

```python
# fill with median

# df['time_since_last_transaction'].fillna(df['time_since_last_transaction'].median(), inplace =True)
# df['time of day'].fillna(df['time of day'].mode()[0], inplace =True)

print(df.isnull().sum())
```

```python
# Step 5: Apply pd.get_dummies function on gender column
    

print(df['gender'].value_counts())
print()
import pandas as pd
```

```python
# Apply dummies function in pandas

df_gender = df['gender']
df_gender = df_gender.astype('category')
df_gender = pd.get_dummies(df_gender)
print(df_gender.shape)
print()

con = [df,df_gender]
df = pd.concat(con, axis=1)
print(df.shape)
print()
df = df.drop('gender',axis=1)
print(df.shape)
print()
```

```python
# Step 6: Apply Label Encoder on City,State,job and trans Column

from sklearn.preprocessing import LabelEncoder

labelencoder_city = LabelEncoder()
labelencoder_state = LabelEncoder()
labelencoder_job = LabelEncoder()
labelencoder_job
```

```python
# preprocessing of the text data

df['city'] = df['city'].str.lower()
df['state'] = df['state'].str.lower()
df['job'] = df['job'].str.lower()



df['city'] = df['city'].str.strip()
df['state'] = df['state'].str.strip()
df['job'] = df['job'].str.strip()


df['city'] = df['city'].str.replace(" ", "_")
df['state'] = df['state'].str.replace(" ", "_")
df['job'] = df['job'].str.replace(" ", "_")


df['city'] = df['city'].str.replace(".", "")
df['state'] = df['state'].str.replace(".", "")
df['job'] = df['job'].str.replace(".", "")


df['city'] = df['city'].str.replace(",", "")
df['state'] = df['state'].str.replace(",", "")
df['job'] = df['job'].str.replace(",", "")


df['city'] = df['city'].str.replace("-", "_")
df['state'] = df['state'].str.replace("-", "_")
df['job'] = df['job'].str.replace("-", "_")


df['city'] = df['city'].str.strip()
df['state'] = df['state'].str.strip()
df['job'] = df['job'].str.strip()

df['city'] = labelencoder_city.fit_transform(df['city'])
df['state'] = labelencoder_state.fit_transform(df['state'])
df['job'] = labelencoder_job.fit_transform(df['job'])

print(df.shape)
print()
```

```python
# Step 7: Splitting the data into training and testing


x = df.drop('is_fraud', axis = 1)
print(x.shape)

y = df['is_fraud']

df['category'] = df['category'].astype('category')
print(x.shape)
print(y.shape)

from sklearn.model_selection import train_test_split

x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.25)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
# Step 8: Apply SMOT Analysis to handle Class imbalance in the target varibale

import sklearn

from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train , y_train = smote.fit_resample(x_train,y_train)
print(y_train.value_counts())
print(x_train.shape)
y_train.shape
```

```python
# Step 9: Apply Standard Scaler

from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler()
standard_scaler

x_train = standard_scaler.fit_transform(x_train)
x_train
x_test = standard_scaler.transform(x_test)
x_test
```

```python
# Step 10:  Apply PCA

from sklearn.decomposition import PCA
pca = PCA(n_components=0.99)

pca.fit(x_train)
x_train = pca.transform(x_train)
x_test = pca.transform(x_test)
print(x_train.shape)
print(x_test.shape)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()

# train the model

model_dt.fit(x_train,y_train)

# model score

print()
print('model training score: ',model_dt.score(x_train,y_train))
print('model testing score: ',model_dt.score(x_test,y_test))
print()

# model prediction

model_prediction_dt = model_dt.predict(x_test)

# model accuracy score

from sklearn.metrics import accuracy_score

print('accuracy score : ',accuracy_score(model_prediction_dt,y_test))
```

# Logisctic Regression Model

```python
# Now, trained the model

from sklearn.linear_model import LogisticRegression

# train model

model_lr = LogisticRegression()
model_lr.fit(x_train,y_train)

# model prediction

model_prediction_lr = model_lr.predict(x_test)
print('Model Predicton: ', model_prediction_lr)
print()
print(y_test[:10])
print()

# model score

print('Model Training Score: ',model_lr.score(x_train,y_train))
print()
print('Model Testing Score: ',model_lr.score(x_test,y_test))
print()

# accuracy score

from sklearn.metrics import accuracy_score

print('accuracy score: ',accuracy_score(model_prediction_lr,y_test))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier()

# train the model

model_rf.fit(x_train,y_train)

# model score

print('model training score: ',model_rf.score(x_train,y_train))
print()
print('model testing score: ',model_rf.score(x_test,y_test))
print()

# model prediction

model_prediction_rf = model_rf.predict(x_test)

# model accuracy score

print('accuracy score : ',accuracy_score(model_prediction_rf,y_test))
```
